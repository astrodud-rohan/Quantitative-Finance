## Visualizing specification sensitivity

```python
fig, ax = plt.subplots(figsize=(8, 5))
x = range(len(sensitivity_df))
width = 0.35
ax.bar([i - width/2 for i in x], sensitivity_df['cumulative_loss_lag1_linear'], width, label='Linear (current)')
ax.bar([i + width/2 for i in x], sensitivity_df['cumulative_loss_lag1_quadratic'], width, label='Quadratic (alternative)')
ax.set_xticks(list(x)); ax.set_xticklabels(sensitivity_df['scenario'])
ax.legend(); ax.set_title('Specification Sensitivity')
plt.show()
```

**Finding:** the linear vs. quadratic specification choice produces a negligible difference in Baseline (+5.1%) and Adverse (-0.8%) projected losses, but a 14.5% difference in Severely Adverse -- specification uncertainty is concentrated exactly where the model is already extrapolating most heavily (per notebook 02), compounding the risk.
