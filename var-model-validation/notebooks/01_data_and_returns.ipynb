# 01 — Data and Returns

Loads the daily equity index price series and inspects the return distribution, volatility clustering, and the stress regime that will later drive the backtesting findings.

```python
import sys
sys.path.append('../src')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../data/raw/index_prices.csv', parse_dates=['date'])
print(df.shape)
df.head()
```

## Price and return series

```python
fig, axes = plt.subplots(2, 1, figsize=(11, 6), sharex=True)
axes[0].plot(df['date'], df['price'])
axes[0].set_title('Index Price')
axes[1].plot(df['date'], df['log_return'], linewidth=0.6)
axes[1].set_title('Daily Log Returns')
plt.tight_layout()
plt.show()
```

## Return distribution statistics

Note the excess kurtosis (fat tails) - this is what causes Parametric (Normal-distribution) VaR to systematically underestimate risk, a theme that recurs throughout the backtesting notebook.

```python
from scipy.stats import kurtosis, skew
print(f'Annualized vol: {df["log_return"].std() * np.sqrt(252):.4f}')
print(f'Skewness: {skew(df["log_return"]):.4f}')
print(f'Excess kurtosis: {kurtosis(df["log_return"]):.4f}  (Normal = 0)')
```

## Volatility regimes

A rolling 20-day realized volatility makes the stress period immediately visible.

```python
df['rolling_vol_20d'] = df['log_return'].rolling(20).std() * np.sqrt(252)
plt.figure(figsize=(11, 4))
plt.plot(df['date'], df['rolling_vol_20d'])
plt.axvspan(pd.Timestamp('2023-06-15'), pd.Timestamp('2023-09-21'), color='orange', alpha=0.15, label='Stress regime')
plt.title('Rolling 20-day Annualized Volatility')
plt.legend()
plt.show()
```
