# 02 — VaR Models

**Wearing the Market Risk Quant hat.** Computes four VaR methodologies: the production champion (Historical Simulation), a naive alternative (Parametric), and two challengers (EWMA, Filtered Historical Simulation).

This notebook builds the models. Backtesting and validation judgment happen independently in `03_backtesting_validation.ipynb`.

```python
import sys
sys.path.append('../src')
import pandas as pd
import matplotlib.pyplot as plt
from var_models import compute_all_var_models
%matplotlib inline
```

```python
df = pd.read_csv('../data/raw/index_prices.csv', parse_dates=['date'])
scored = compute_all_var_models(df, window=250, confidence=0.99)
scored.dropna(subset=['hs_var']).tail()
```

## Visual comparison of all four VaR estimates

```python
plot_df = scored.dropna(subset=['hs_var'])
plt.figure(figsize=(12, 5))
for col, label in [('hs_var','HS (Champion)'), ('parametric_var','Parametric'),
                   ('ewma_var','EWMA'), ('fhs_var','FHS (Challenger)')]:
    plt.plot(plot_df['date'], plot_df[col], label=label, linewidth=1)
plt.axvspan(pd.Timestamp('2023-06-15'), pd.Timestamp('2023-09-21'), color='orange', alpha=0.1)
plt.title('VaR Estimates by Model')
plt.legend()
plt.show()
```

Notice how HS and Parametric are slow to react to the stress regime and stay elevated long after it ends, while EWMA and FHS both rise and fall much more quickly — this is the core mechanical difference that will drive the backtesting results in the next notebook.
